```python
import numpy as np
```

#### ***1. Đọc gold_price_hcm.csv và gold_price_hn.csv bằng np.loadtxt hoặc np.genfromtxt.***

```python
hcm = np.genfromtxt("gold_price_hcm.csv", delimiter=",", dtype="U20,f8", skip_header=1)
hn  = np.genfromtxt("gold_price_hanoi.csv", delimiter=",", dtype="U20,f8", skip_header=1)

print("=== 1. Đọc file xong ===")
print("HCM sample:", hcm[:3])
print("HN  sample:", hn[:3])
print()
```

#### ***2. Tách ngày và giá thành các mảng riêng.***

```python
date_hcm = hcm["f0"]
price_hcm = hcm["f1"]
date_hn = hn["f0"]
price_hn = hn["f1"]

print("=== 2. Tách ngày & giá ===")
print("date_hcm:", date_hcm[:5])
print("price_hcm:", price_hcm[:5])
print()
```

#### ***3. Chuyển cột ngày sang datetime64[D].***

```python
date_hcm = date_hcm.astype("datetime64[D]")
date_hn  = date_hn.astype("datetime64[D]")

print("=== 3. Chuyển ngày datetime ===")
print("date_hcm dtype:", date_hcm.dtype)
print()
```

#### ***4. Kiểm tra dtype, shape, in 5 giá trị đầu và cuối.***

```python
print("=== 4. Kiểm tra dtype & shape ===")
print("HCM shape:", hcm.shape, "HN shape:", hn.shape)
print("HCM head:", price_hcm[:5])
print("HN head :", price_hn[:5])
print("HCM tail:", price_hcm[-5:])
print("HN tail :", price_hn[-5:])
print("HCM dtype:", hcm.dtype)
print("HN  dtype:", hn.dtype)

print()
```

#### ***5. Slicing: lấy 7 ngày đầu, 7 ngày giữa và 7 ngày cuối***

```python
# === SLICING TP.HCM ===
first_7 = price_hcm[:7]
mid_7   = price_hcm[len(price_hcm)//2 - 3 : len(price_hcm)//2 + 4]
last_7  = price_hcm[-7:]

print("=== 5. Slicing 7 ngày (HCM) ===")
print("7 ngày đầu:", first_7)
print("7 ngày giữa:", mid_7)
print("7 ngày cuối:", last_7)
print()

# === SLICING HÀ NỘI ===
first_7_hn = price_hn[:7]
mid_7_hn   = price_hn[len(price_hn)//2 - 3 : len(price_hn)//2 + 4]
last_7_hn  = price_hn[-7:]

print("=== 5. Slicing 7 ngày (HÀ NỘI) ===")
print("7 ngày đầu:", first_7_hn)
print("7 ngày giữa:", mid_7_hn)
print("7 ngày cuối:", last_7_hn)
print()
```

#### ***6. Boolean mask: tìm ngày TP.HCM cao hơn Hà Nội.***

```python
mask_higher = price_hcm > price_hn
days_higher = date_hcm[mask_higher]

print("=== 6. HCM cao hơn Hà Nội ===")
print("Số ngày:", len(days_higher))
print("Các ngày:", days_higher)
print()
```

#### ***7. Fancy indexing: tìm top-5 ngày giá cao nhất mỗi khu vực.***

```python
top5_hcm_idx = np.argsort(price_hcm)[-5:]
top5_hn_idx  = np.argsort(price_hn)[-5:]

print("=== 7. Top 5 ngày giá cao nhất HCM ===")
for i in top5_hcm_idx:
    print(date_hcm[i], price_hcm[i])
print()

print("=== 7. Top 5 ngày giá cao nhất HN ===")
for i in top5_hn_idx:
    print(date_hn[i], price_hn[i])
print()
```

#### ***8. Broadcasting: tính chênh lệch giá theo từng ngày***

```python
diff = price_hcm - price_hn
print("=== 8. Chênh lệch giá (HCM - HN) ===")
print(diff[:10])
print("Chênh lệch trung bình:", np.mean(diff))
print("Chênh lệch max:", np.max(diff))
print("Chênh lệch min:", np.min(diff))
print()
```

#### ***9. Dùng np.diff để phân tích biến động ngày liền kề.***

```python
daily_change_hcm = np.diff(price_hcm)
daily_change_hn  = np.diff(price_hn)

print("=== 9. Biến động ngày liền kề ===")
print("HCM diff:", daily_change_hcm[:10])
print("HN diff :", daily_change_hn[:10])
print()
```

#### ***10. Tính mean, std, max, min, argmax, argmin.***

```python
print("=== 10. Thống kê cơ bản ===")
print("HCM mean:", np.mean(price_hcm))
print("HN  mean:", np.mean(price_hn))
print("HCM std:", np.std(price_hcm))
print("HN  std:", np.std(price_hn))
print("HCM max:", np.max(price_hcm), "tại ngày", date_hcm[np.argmax(price_hcm)])
print("HN  max:", np.max(price_hn),  "tại ngày", date_hn[np.argmax(price_hn)])
print("HCM min:", np.min(price_hcm), "tại ngày", date_hcm[np.argmin(price_hcm)])
print("HN  min:", np.min(price_hn),  "tại ngày", date_hn[np.argmin(price_hn)])
print()
```

#### ***11. Tạo bảng tổng hợp và lưu ket_qua_case_1.csv bằng np.savetxt.***

```python
summary = np.column_stack((
    date_hcm.astype("str"),
    price_hcm,
    price_hn,
    diff
))

np.savetxt(
    "ket_qua_case_1.csv",
    summary,
    delimiter=",",
    fmt="%s",
    header="date,price_hcm,price_hn,chenh_lech",
    comments=""
)

print("=== 11. Đã lưu file ket_qua_case_1.csv ===")
```

#### ***12. (Tuỳ chọn) Vẽ biểu đồ line hoặc bar minh hoạ.***

```python
import numpy as np
import matplotlib.pyplot as plt

```

****Line Chart: Biến động giá vàng TP.HCM vs Hà Nội****

```python
plt.figure(figsize=(12,6))
plt.plot(date_hcm, price_hcm, marker='o', label='TP.HCM', color='gold', linewidth=2)
plt.plot(date_hn, price_hn, marker='o', label='Hà Nội', color='orange', linewidth=2)
plt.title("Biến động giá vàng TP.HCM vs Hà Nội (6 tháng)")
plt.xlabel("Ngày")
plt.ylabel("Giá vàng")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("case1_line_chart.png", dpi=150) 
plt.show()
```

****Bar Chart: Chênh lệch giá vàng HCM - HN****

```python
plt.figure(figsize=(12,5))
plt.bar(date_hcm, diff, color='skyblue')
plt.axhline(0, color='black', linestyle='--')  
plt.title("Chênh lệch giá vàng TP.HCM - Hà Nội")
plt.xlabel("Ngày")
plt.ylabel("Chênh lệch giá (HCM - HN)")
plt.tight_layout()
plt.savefig("case1_bar_chart.png", dpi=150)  
plt.show()
```

****Line Chart: Biến động ngày liền kề (np.diff)****

```python
diff_hcm = np.diff(price_hcm)
diff_hn  = np.diff(price_hn)

plt.figure(figsize=(12,5))
plt.plot(date_hcm[1:], diff_hcm, label="Biến động TP.HCM", color='green')
plt.plot(date_hn[1:], diff_hn, label="Biến động Hà Nội", color='red')
plt.title("Biến động giá vàng theo từng ngày (np.diff)")
plt.xlabel("Ngày")
plt.ylabel("Mức tăng/giảm")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("case1_diff_chart.png", dpi=150)  
plt.show()
```
